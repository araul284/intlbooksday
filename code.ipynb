```python
import requests, pandas as pd

# simple Ergast results fetcher for a season
def fetch_season_results(season):
    url = f'http://ergast.com/api/f1/{season}/results.json?limit=1000'
    r = requests.get(url).json()
    rows = []
    for race in r['MRData']['RaceTable']['Races']:
        race_date = race['date']
        round_ = race['round']
        season_ = race['season']
        circuit = race['Circuit']['circuitName']
        for res in race['Results']:
            rows.append({
                'season': season_,
                'round': round_,
                'race_date': race_date,
                'circuit': circuit,
                'driver_id': res['Driver']['driverId'],
                'driver_name': res['Driver']['familyName'],
                'constructor': res['Constructor']['name'],
                'grid': int(res['grid']),
                'position': int(res['position']) if 'position' in res and res['position'].isdigit() else None,
            })
    return pd.DataFrame(rows)

df = fetch_season_results(2022)
df.head()
```
